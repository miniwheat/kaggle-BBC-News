```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
m = np.array([
    [315, 1, 7, 1, 12],
    [3, 223, 6, 2, 39],
    [17, 0, 240, 2, 15],
    [1, 2, 0, 343, 0],
    [0, 4, 0, 4, 253]
])

labels = ['sports', 'politics', 'bus', 'ent', 'tech']
```

```python
# heatmap basd on the counts

fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(m, cmap='rocket_r', annot=True, fmt='', annot_kws={'size': 16}, xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted\n\n(cell % represents Accuracy)', fontsize=16)
plt.ylabel('True', fontsize=16)
plt.title('Confusion Matrix', fontsize=20);
```

```python
# heatmap basd on the accuracy

fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(m/np.sum(m, axis=1), cmap='rocket_r', annot=True, fmt='.2%', annot_kws={'size': 16}, xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted\n\n(cell % represents Accuracy)', fontsize=16)
plt.ylabel('True', fontsize=16)
plt.title('Confusion Matrix', fontsize=20);
```

```python
# Compare with slightly different matrix

m2 = np.array([
    [310,   1,  16,   0,   9],
    [  2, 259,   7,   0,   5],
    [  1,   1, 269,   1,   2],
    [  0,   2,   0, 344,   0],
    [  6,   4,   7,   1, 243]
])
```

```python
# heatmap basd on the accuracy - with custom annotations

counts = m2.flatten()
p = (m2/np.sum(m2, axis=1)).flatten()
accuracy = p[::6]

annots = [str(count) for count in counts]
for i in range(len(accuracy)):
    annots[i*6] = annots[i*6] + '\n' + f'{accuracy[i]:.2%}'

annots = np.asarray(annots).reshape(5,5)

fig2, ax = plt.subplots(figsize=(12,8))
sns.heatmap(m2/np.sum(m2, axis=1), cmap='rocket_r', annot=annots, fmt='', annot_kws={'size': 16}, xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted\n\n(cell % represents Accuracy)', fontsize=16)
plt.ylabel('True', fontsize=16)
plt.title('Confusion Matrix', fontsize=20);
```

# Show the existing figures side by side

```python
# Compare the figures side by side

fig, axs = plt.subplots(1,2, figsize = (20, 8))
fig.suptitle('Confusion Matrix Comparison', fontsize=20)

counts = m.flatten()
p = (m/np.sum(m, axis=1)).flatten()
accuracy = p[::6]
annots = [str(count) for count in counts]
for i in range(len(accuracy)):
    annots[i*6] = annots[i*6] + '\n' + f'{accuracy[i]:.2%}'
annots = np.asarray(annots).reshape(5,5)

axs[0].set_title('Initial')
sns.heatmap(m/np.sum(m, axis=1), cmap='rocket_r', annot=annots, fmt='', annot_kws={'size': 14}, xticklabels=labels, yticklabels=labels, ax=axs[0])

counts = m2.flatten()
p = (m2/np.sum(m2, axis=1)).flatten()
accuracy = p[::6]
annots = [str(count) for count in counts]
for i in range(len(accuracy)):
    annots[i*6] = annots[i*6] + '\n' + f'{accuracy[i]:.2%}'
annots = np.asarray(annots).reshape(5,5)

axs[1].set_title('Parameter Tuned')
sns.heatmap(m2/np.sum(m2, axis=1), cmap='rocket_r', annot=annots, fmt='', annot_kws={'size': 14}, xticklabels=labels, yticklabels=labels, ax=axs[1]);
```
